```python
model_path = 'runs/detect/train3/weights/best.pt'  # Update with the correct path to the best weights
model = YOLO(model_path)
```

```python
test_images_dir = '/content/drive/MyDrive/aadhar_test'

# Get a list of all image files in the directory
image_files = [f for f in os.listdir(test_images_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Run inference on each image and display results
for image_file in image_files:
    image_path = os.path.join(test_images_dir, image_file)  # Full path to the image
    results = model(image_path)  # Run inference

    # Display results for each image
    for result in results:
        result.show()  # Display the image with predictions
```

```python
for image_file in image_files:
    image_path = os.path.join(test_images_dir, image_file)  # Full path to the image
    results = model(image_path)

    for result in results:
        boxes = result.boxes  # Bounding boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].tolist()  # Get the coordinates of the bounding box
            confidence = box.conf[0].item()  # Get the confidence score
            cls = int(box.cls[0].item())  # Get the class index

            print(f"Image: {image_file}, Detected Class: {cls}, Confidence: {confidence:.2f}, Coordinates: ({x1}, {y1}), ({x2}, {y2})")
```

```python
import os
import cv2
import pytesseract
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Load the original image again for cropping and OCR
for image_file in image_files:
    image_path = os.path.join(test_images_dir, image_file)  # Full path to the image
    image = cv2.imread(image_path)

    results = model(image_path)
    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].tolist()  # Get the coordinates of the bounding box
            cropped_img = image[int(y1):int(y2), int(x1):int(x2)]  # Crop the image

            # Use Tesseract to extract text from the cropped image
            text = pytesseract.image_to_string(cropped_img)
            print(f"Image: {image_file}, Extracted Text: {text.strip()}")
```
